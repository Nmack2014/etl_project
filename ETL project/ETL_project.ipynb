```python
df_2020.rename(columns={"total_votes": "total_votes_2020"})
```

```python
df_2016 = pd.read_csv("nytimes_presidential_elections_2016_results_county.csv")
```

```python
df_2016
```

```python
sum_column = df_2016["Clinton"] + df_2016["Trump"]
df_2016["total_votes_2016"] = sum_column
```

```python
df_2016['county'] = df_2016['Place'].astype(str) + " County"
```

```python
df_2016
```

```python
merged_df = pd.merge(df_2016, df_2020,  how='inner', left_on=['State','county'], right_on = ['state','county'])
```

```python
merged_df
```

```python
merged_df = merged_df.drop(columns=['Clinton', 'Trump', 'Rpt', 'Vote by county', 'Vote by town', 'Place', 'State', 'total_votes', 'percent'])
```

```python
merged_df = merged_df.rename(columns={"current_votes": "total_votes_2020"})
```

```python
merged_df.head()
```

```python
merged_df['percent change'] = ((merged_df['total_votes_2020'] - merged_df['total_votes_2016']) / merged_df['total_votes_2016']) * 100
```

```python
merged_df.head()
```

```python
merged_df.loc[merged_df['county'] == 'Cook County']
```

```python
illinois_df = merged_df.loc[merged_df['state'] == 'Illinois']
```

```python
illinois_df.head(50)
```

```python
illinois_graph = illinois_df.plot.bar(x='county', y='percent change', rot=0, xticks=None)
illinois_graph
```

```python
lowest_change = illinois_df.loc[illinois_df['percent change'] == illinois_df['percent change'].min()]
lowest_change
```

```python
highest_change = illinois_df.loc[illinois_df['percent change'] == illinois_df['percent change'].max()]
highest_change
```
